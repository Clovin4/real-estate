```python
df = pd.read_csv('../data/raw/Metro_invt_fs_uc_sfrcondo_sm_month.csv')
df.shape
```

```python
df.head()
```

```python
# transpose the dataframe
df = df.T
```

```python
# assign the Region Name row as the column names\
df.columns = df.iloc[2]
```

```python
# create a new dataframe with only the dated rows
df = df.iloc[5:]
```

```python
df
```

```python
# set the index to be the date
df.index = pd.to_datetime(df.index)
```

```python
df
```

```python
df.index
```

```python
# plot the data for the united states column
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df['United States'], mode='lines'))
fig.update_layout(title='United States Inventory of Homes for Sale',
                    xaxis_title='Date',
                    yaxis_title='Inventory')
fig.show()
```

```python
# drop the united states column
df = df.drop(columns=['United States'])
```

```python
# plot the data for all columns using a dropdown menu for each column
fig = go.Figure()
for column in df.columns:
    fig.add_trace(go.Scatter(x=df.index, y=df[column], mode='lines', name=column))
fig.show()
```
